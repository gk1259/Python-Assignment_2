1. Write a program to accept percentage from the user and display the grade according to the following criteria:



```python
percentage = int(input('your percentage'))

if percentage >90 :
    print('full marks' , (percentage * 5))
    print('you will get grade A')
elif percentage >80 and percentage <=90 :
    print('full marks' , (percentage * 5))
    print('you will get grade B')
elif percentage >=60 and percentage <= 80 :
    print('full marks' , (percentage * 5))
    print('you will get grade C')
else :
    print('full marks' , (percentage * 5))
    print('you will get grade D')  
```

2. Write a program to accept the cost price of a bike and display the road tax to be paid according to the following criteria:



```python
bike_cost = int(input('bike cost'))

if bike_cost >100000 :
    print('you will pay tax' , '15%')
    print('your tax amount is' , int((bike_cost * 15/100)))
    
elif bike_cost >50000 and bike_cost <= 100000 :
    print('you will pay tax' , '10%')
    print('your tax amount is' , int((bike_cost * 10/100)))
    
else :
    print('you will pay tax' , '5%')
    print('your tax amount is' , int((bike_cost * 5/100)))
```

4. Check how many times a given number can be divided by 3 before it is less than or equal to 10.

```python
num = int(input("Enter a number"))
count = 0
while num > 10:
    num = num / 3
    count = count + 1
    
print(f"{count}" , 'times')
```

6. Use nested while loop to print 3 different pattern.

According to the question,three different patterns are as follows :

```python
#Pattern 1

n = int(input('enter your row'))

while n <= 10 :
    k = 0
    while k <= n :
        print(k , end = " ")
        k = k + 1
    print(" ")
    n = n + 1
```

```python
#Pattern 2

n = int(input('enter your row'))
i = 1
while i <= n :
    j = i
    while j < n :
        print(' ' , end = ' ')
        j = j + 1
    k = 1
    while k <= i :
        print('*' , end = ' ')
        k = k + 1
    print()
    i = i + 1
```

```python
#Pattern 3

n = int(input('Enter the row:'))
i = 0
while i <= n - 1 :
    j = 0
    while j < i :
       
        print('', end = ' ')
        j += 1
    k = i
    while k <= n - 1 :
        print('+', end = ' ')
        k += 1
    print()
    i += 1
```

7. Reverse a while loop to display numbers from 10 to 1.
